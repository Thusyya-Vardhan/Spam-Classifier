# SMS Spam Detector
## Text Preprocessing & Feature Engineering

**Goal:** Clean raw text data and convert it into numerical features that our ML model can understand.

## 1. Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

nltk.download('stopwords')
```

## 2. Loading Cleaned Dataset

```python
df = pd.read_csv('../data/spam_clean.csv')
print(df.shape)
df.head()
```

## 3. Text Preprocessing

Raw text contains a lot of noise — punctuation, numbers, stopwords (like "the", "is", "a") that add no value for spam detection. We need to clean it before feeding it to our model.

Steps:
1. Lowercase the text
2. Remove punctuation and numbers
3. Remove stopwords
4. Stemming — reduce words to their root form (e.g. "calling", "called" → "call")

```python
ps = PorterStemmer()

def preprocess(text):
    # 1. Lowercase
    text = text.lower()
    # 2. Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    # 3. Tokenize
    tokens = text.split()
    # 4. Remove stopwords and stem
    tokens = [ps.stem(w) for w in tokens if w not in stopwords.words('english')]
    return ' '.join(tokens)
```

```python
# Test it on a sample message
sample = "Congratulations! You have WON a FREE prize. Call now at 98765!"
print("Before:", sample)
print("After:", preprocess(sample))
```

## 4. Applying Preprocessing to Entire Dataset

```python
df['processed'] = df['message'].apply(preprocess)
print("Preprocessing complete!")
df[['message', 'processed']].head(10)
```

## 5. Saving Preprocessed Data

```python
df.to_csv('../data/spam_processed.csv', index=False)
print("Processed data saved!")
```
